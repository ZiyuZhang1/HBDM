```python
root = r'D:\study\thesis\project\HBDM-main\ppi_results\clusters'

# Get a list of all folders in the root directory
folders = [0,10,30,70,150,300,558]
# folders = [27,7]
max_y = 0  # Variable to store the maximum y-axis value
subplotname = ['Cellular component', 'Molecular function','Biological process','KEGG pathways', 'Reactom pathways']
for foldername in folders:
    folderpath = os.path.join(root, str(foldername))
    files = ['enrichment.Component.tsv','enrichment.Function.tsv', 'enrichment.Process.tsv','enrichment.KEGG.tsv','enrichment.RCTM.tsv']

    # Create a subplot for each file in the folder
    fig, axs = plt.subplots(1, len(files), figsize=(15, 5))

    for i, file in enumerate(files):
        df = pd.read_csv(os.path.join(folderpath, file), sep='\t')
        df = df[df['false discovery rate'] < 0.01]
        histogram, bins, _ = axs[i].hist(df['observed gene count']/df['background gene count'])
        axs[i].set_xlabel('observed/background')
        axs[i].set_ylabel('count')
        axs[i].set_title(subplotname[i])

        max_y = max(max_y, max(histogram))  # Update max_y

    # Set the y-axis limits to be the same for all subplots
    for i in range(len(files)):
        axs[i].set_ylim(0, max_y)

    plt.suptitle('Plots for Cluster ' + str(foldername))
    plt.tight_layout()
    plt.show()
```

```python
import os
import matplotlib.pyplot as plt

root = r'D:\study\thesis\project\HBDM-main\ppi_results\clusters'

# Get a list of all folders in the root directory
folders = [558]

max_y = 0  # Variable to store the maximum y-axis value
subplotname = ['Cellular component', 'Biological process']

for foldername in folders:
    folderpath = os.path.join(root, str(foldername))
    files = ['enrichment.Component.tsv', 'enrichment.Process.tsv']

    # Check if all files are present in the folder
    if all(os.path.exists(os.path.join(folderpath, file)) for file in files):
        # Create a subplot for each file in the folder
        fig, axs = plt.subplots(1, len(files), figsize=(15, 5))

        for i, file in enumerate(files):
            df = pd.read_csv(os.path.join(folderpath, file), sep='\t')
            df = df[df['false discovery rate'] < 0.01]
            histogram, bins, _ = axs[i].hist(df['strength'].tolist())
            axs[i].set_xlabel('strength')
            axs[i].set_ylabel('count')
            axs[i].set_title(subplotname[i])

            max_y = max(max_y, max(histogram))  # Update max_y

        # Set the y-axis limits to be the same for all subplots
        for i in range(len(files)):
            axs[i].set_ylim(0, max_y)

        plt.suptitle('Plots for Cluster ' + str(foldername))
        plt.tight_layout()
        plt.show()
    else:
        print(f"No files found in folder {foldername}. No plots created.")
```
