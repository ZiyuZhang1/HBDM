```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
sp_d = pd.read_csv(r'D:\study\thesis\project\HBDM-main\ppi_results\test_results\complexes\sp_detailed_top20.csv')
sp_10 = pd.read_csv(r'D:\study\thesis\project\HBDM-main\ppi_results\test_results\complexes\sp_10level_top50.csv')
T_T_D4 = pd.read_csv(r'D:\study\thesis\project\HBDM-main\ppi_results\test_results\complexes\T-T-D4-50.csv')
```

```python
sp_d_p = sp_d[['precision_20', 'coverage_20']].add_prefix('sp_d_')
sp_10_p = sp_10[['precision_50', 'coverage_50']].add_prefix('sp_10_')
sp_d_p['complex'] = sp_d['complex']
sp_10_p['complex'] = sp_10['complex']
```

```python
sp = sp_d_p.merge(sp_10_p)
T_T_D4 = T_T_D4.merge(sp)
```

```python
columns_to_check = ['precision_20', 'precision_50', 'coverage_20', 'coverage_50',
                    'sp_d_precision_20', 'sp_d_coverage_20', 'sp_10_precision_50', 'sp_10_coverage_50']

# Apply a filter to keep rows where the sum of values in the specified columns is not equal to 0
T_T_D4 = T_T_D4[T_T_D4[columns_to_check].sum(axis=1) != 0]
```

```python
T_T_D4.columns
```

```python


# Your existing code

df = T_T_D4

fig = plt.figure(figsize=(5, 4))  # Adjust the figsize as needed

x = ['precision_top20', 'precision_top50']
barWidth = 0.25
r1 = np.arange(len(x))
r2 = [z + barWidth for z in r1]

# Calculate the means for the two categories
mean_hbdm_precision_20 = [df['precision_20'].mean(), df['precision_50'].mean()]
mean_sp_precision_20 = [df['sp_d_precision_20'].mean(), df['sp_10_precision_50'].mean()]

plt.bar(r1, mean_hbdm_precision_20, width=barWidth, label='HBDM_precision')
plt.bar(r2, mean_sp_precision_20, width=barWidth, label='SP_precision')

# Set the x-axis ticks and labels
plt.xticks([r + barWidth / 2 for r in r1], x)

# Add the legend to the plot
plt.legend()

# Additional code for labeling the axes, title, etc.
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Precision')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

df = T_T_D4  # Assuming T_T_D4 is your DataFrame

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Data
x = ['HBDM', 'SP_d']
barWidth = 0.25
r = np.arange(len(x))

# Calculate the means and variances for precision_top20
mean_precision_hbdm = df['precision_20'].mean()
variance_precision_hbdm = df['precision_20'].var()
mean_precision_sp = df['sp_d_precision_20'].mean()
variance_precision_sp = df['sp_d_precision_20'].var()

# Calculate the means and variances for coverage_top20
mean_coverage_hbdm = df['coverage_20'].mean()
variance_coverage_hbdm = df['coverage_20'].var()
mean_coverage_sp = df['sp_d_coverage_20'].mean()
variance_coverage_sp = df['sp_d_coverage_20'].var()

# Create subplots for precision
ax1.bar(r, [mean_precision_hbdm, mean_precision_sp], width=barWidth, label='Precision', yerr=[variance_precision_hbdm, variance_precision_sp], capsize=7)
ax1.set_xticks(r)
ax1.set_xticklabels(x)
ax1.set_ylabel('Percentage')
ax1.set_title('Precision - HDBM vs SP_d')
ax1.legend()

# Create subplots for coverage
ax2.bar(r, [mean_coverage_hbdm, mean_coverage_sp], width=barWidth, label='Coverage', yerr=[variance_coverage_hbdm, variance_coverage_sp], capsize=7)
ax2.set_xticks(r)
ax2.set_xticklabels(x)
ax2.set_ylabel('Percentage')
ax2.set_title('Coverage - HDBM vs SP_d')
ax2.legend()

# Adjust the layout to prevent subplot overlapping
plt.tight_layout()

plt.show()
```
